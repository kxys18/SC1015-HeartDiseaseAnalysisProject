## Naive Bayes Classifier
we use the Naive Bayes Classifier since from EDA we observed that most of the variables in the data set are independent, which meets the condition of the Naive Bayes Method. Also, it usually performs well with categorized data. Since it is based on Bayes's theorem so that it has a reliable theoretical foundation.

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() 
from scipy.stats import ttest_ind
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('CVD_cleaned.csv')
print(df.head())
```

```python


# 检查列名是否存在于数据框中
print('General_Health' in df.columns)

print(df.columns)
```

```python
#LabelEncoder
label_encoder = LabelEncoder()


df['General_Health'] = label_encoder.fit_transform(df['General_Health'])
df['Checkup'] = label_encoder.fit_transform(df['Checkup'])
df['Exercise'] = label_encoder.fit_transform(df['Exercise'])
df['Skin_Cancer'] = label_encoder.fit_transform(df['Skin_Cancer'])
df['Other_Cancer'] = label_encoder.fit_transform(df['Other_Cancer'])
df['Depression'] = label_encoder.fit_transform(df['Depression'])
df['Diabetes'] = label_encoder.fit_transform(df['Diabetes'])
df['Arthritis'] = label_encoder.fit_transform(df['Arthritis'])
df['Sex'] = label_encoder.fit_transform(df['Sex'])
df['Age_Category'] = label_encoder.fit_transform(df['Age_Category'])
df['Heart_Disease'] = label_encoder.fit_transform(df['Heart_Disease'])
df['Smoking_History'] = label_encoder.fit_transform(df['Smoking_History'])
```

```python
df.head()
```

```python

X = df.drop('Heart_Disease', axis=1)
y = df['Heart_Disease']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Initialize
nb_model = GaussianNB()

# Train
nb_model.fit(X_train, y_train)

# Pridict the test cases
y_pred = nb_model.predict(X_test)
```

```python
# Calculate classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report:\n", class_report)

# Calculate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

The performance of the Naive Bayes Classifier is not very satisfying despite that the overall accuracy is high. But in reality, we are requiring the classifier is able to indentify most of the patients. This is possibly because of the imbalance of the data set. We might need to try other models.
